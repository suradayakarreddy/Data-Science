### Support Vector Machine(SVM)

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.model_selection import KFold
```

```python

x = data2.drop('class', axis=1)
y = data2['class']
x=x.values
y=y.values
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

#### Kerner=Linear and radial basis

```python
clf=SVC()
param_grid=[{'kernel':['linear','rbf'],'gamma':[0.5,0.1,0.01],'C':[10,0.1,0.001,0.0001]}]
```

```python
gsv=RandomizedSearchCV(clf,param_grid,cv=5,verbose=300)
gsv.fit(x_train,y_train)
```

```python
gsv.best_params_,gsv.best_score_
```

```python
y_pred=gsv.predict(x_test)
print(classification_report(y_test,y_pred))
```

#### Kernel=Linear

```python
clf1=SVC(kernel='linear',C=10,gamma=0.01)
clf1.fit(x,y)
```

```python
y_pred2=clf1.predict(x_test)
print(classification_report(y_test,y_pred2))
```

#### Kernel=Ploynominal

```python
clf2=SVC()
param_grid_poly=[{'kernel':['poly'],'gamma':[0.5,0.1,0.01],'C':[10,0.1,0.001,0.0001]}]
```

```python
gsv_poly=RandomizedSearchCV(clf2,param_grid_poly,cv=5,verbose=300)
gsv_poly.fit(x_train,y_train)
```

#### Kernel=radial basis

```python
clf3=SVC()
param_grid_rbf=[{'kernel':['rbf'],'gamma':[0.5,0.1,0.01],'C':[10,0.1,0.001,0.0001]}]
```

```python
gsv_rbf=RandomizedSearchCV(clf3,param_grid_rbf,cv=5,verbose=300)
gsv_rbf.fit(x_train,y_train)
```

The Mushroom dataset, while not extremely high-dimensional, contains numerous categorical features. SVMs can handle high-dimensional spaces well, and when using kernel tricks (like the RBF kernel), they can create complex decision boundaries that might capture the intricacies in the data.

 If EDA shows a lot of categorical features that can be effectively transformed into high-dimensional space, SVM’s ability to handle these can be advantageous.

SVMs are less prone to overfitting, especially in high-dimensional spaces, due to the margin maximization principle. This is particularly useful if the dataset shows signs of being linearly separable or nearly separable

If the data has complex, non-linear relationships, an SVM with a linear kernel might not perform well. Although kernel methods can help, they add computational complexity.

If EDA and visualization reveal that the classes are not easily separable by a straight line or hyperplane, then the SVM’s performance might degrade unless a non-linear kernel is used.

 SVMs can be computationally expensive, particularly with large datasets or complex kernels. While the Mushroom dataset isn’t huge, computational constraints can still be a concern for large-scale applications.
