## Build the three diffrent models 

### Simple linear Regression

```python
x=data4_clean.iloc[:,1:]
y=data4_clean.iloc[:,:1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
simplemodel=LinearRegression()
simplemodel.fit(x_train,y_train)
```

```python
y_pred=simplemodel.predict(x_test)
```

```python
mse=mean_squared_error(y_test,y_pred)
r2_1=r2_score(y_test,y_pred)
```

```python
print("Linear Regression Model")
print("Coeffieciants:",simplemodel.coef_)
print("Mean Squared error:",mse)
print("R^2 Score:",r2_1)
```

### Polynominal Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(x_train)
```

```python
polymodel = LinearRegression().fit(X_poly, y_train)
```

```python
X_test_poly = poly.transform(x_test)
y_pred2 = polymodel.predict(X_test_poly)
```

```python
mse2 = mean_squared_error(y_test, y_pred2)
r2_2 = r2_score(y_test, y_pred2)
```

```python
print(" Polynomial Regression Model")
print("Coefficients:", polymodel.coef_)
print("Mean Squared Error:", mse2)
print("R^2 Score:", r2_2)
```

### Multiple Liner Regression

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)
```

```python
multimodel3 = LinearRegression().fit(X_train_scaled, y_train)
```

```python
y_pred3 =multimodel3.predict(X_test_scaled)
```

```python
mse3 = mean_squared_error(y_test, y_pred3)
r2_3 = r2_score(y_test, y_pred3)
```

```python
print(" Multiple Linear Regression Model")
print("Coefficients:",multimodel3.coef_)
print("Mean Squared Error:", mse3)
print("R^2 Score:", r2_3)
```

### Lasso Regression

```python
from sklearn.linear_model import Lasso
```

```python
lasso = Lasso(alpha=0.1) 
lasso.fit(X_train_scaled, y_train)
y_pred_lasso = lasso.predict(X_test_scaled)
```

```python
mse_lasso = mean_squared_error(y_test, y_pred_lasso)
r2_lasso = r2_score(y_test, y_pred_lasso)
```

```python
print("Lasso Regression")
print("Coefficients:", lasso.coef_)
print("Mean Squared Error:", mse_lasso)
print("R^2 Score:", r2_lasso)
```

### Ridge Regression

```python
from sklearn.linear_model import Ridge
```

```python
ridge = Ridge(alpha=1.0)  
ridge.fit(X_train_scaled, y_train)
y_pred_ridge = ridge.predict(X_test_scaled)
```

```python
mse_ridge = mean_squared_error(y_test, y_pred_ridge)
r2_ridge = r2_score(y_test, y_pred_ridge)
```

```python
print("Ridge Regression")
print("Coefficients:", ridge.coef_)
print("Mean Squared Error:", mse_ridge)
print("R^2 Score:", r2_ridge)
```

# Interview Quations

What is Normalization & Standardization and how is it helpful?

Normalization:
Definition: Normalization is the process of scaling numerical data to a standard range, usually [0, 1].
How it works: It adjusts values based on the minimum and maximum of the feature, transforming them proportionally.
Helpfulness: Helps algorithms converge faster, ensures equal influence of features, and makes data easier to interpret.
Standardization:
Definition: Standardization transforms data to have a mean of 0 and a standard deviation of 1.
How it works: It centers the data around 0 and adjusts its spread.
Helpfulness: Improves performance of models that assume normality, enhances interpretability of coefficients, and supports algorithms sensitive to feature scales.
Summary of Benefits:
Both: Improve model performance, facilitate algorithm convergence, and enhance interpretability of results.
Normalization: Useful when features have varying scales and need to be scaled to a specific range.
Standardization: Useful when features should have mean 0 and unit variance, especially in algorithms assuming Gaussian (normal) distribution.

What techniques can be used to address multicollinearity in multiple linear regression?

Feature Selection: Choose only the most relevant features that have a stronger relationship with the target variable.

Principal Component Analysis (PCA): Transform correlated variables into a smaller set of uncorrelated components.

Regularization Techniques: Use methods like Ridge regression, which penalizes large coefficients, reducing their impact.
